```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)
```

```python
import numpy as np
x = np.random.randint(0, 10, (4,1))
y = 2 * x
print(x)
print(y)
```

```python
import numpy as np
W1 = np.random.rand(1,1)
b1 = np.random.rand(1,1)
W2 = np.random.rand(1,1)
b2 = np.random.rand(1,1)
print(W1, b1, W2, b2)
```

```python
z1 = x * W1 + b1
```

```python
print("Hidden layer input z1:", z1)
```

```python
a1 = sigmoid(z1)
print("Hidden layer activation a1=\n", a1)
```

```python
z2 = a1 * W2 + b2
y_pred = sigmoid(z2)
print("Predicted output y=\n", y_pred)
```

```python
mse = np.mean((y - y_pred)**2)
print("MSE:", mse)
```

```python
loss = []
loss.append(np.mean((y - y_pred)**2))
```

```python
error = y - y_pred
print("Error:\n", error)
delta_output = error * y_pred * (1 - y_pred)
print("Delta output:\n", delta_output)
delta_hidden = delta_output * W2 * a1 * (1 - a1)
print("Delta hidden:\n", delta_hidden)
```

```python
lr = 0.1
W2 = W2 + lr * a1 * delta_output
b2 = b2 + lr * delta_output
W1 = W1 + lr * x * delta_hidden
b1 = b1 + lr * delta_hidden
```

```python
epochs = 5

for i in range(epochs):
    z1 = x * W1 + b1
    a1 = sigmoid(z1)
    z2 = a1 * W2 + b2
    y_pred = sigmoid(z2)
    mse = np.mean((y - y_pred)**2)
    loss.append(mse)
    error = y - y_pred
    delta_output = error * y_pred * (1 - y_pred)
    delta_hidden = delta_output * W2 * a1 * (1 - a1)
    W2 = W2 + lr * a1 * delta_output
    b2 = b2 + lr * delta_output
    W1 = W1 + lr * x * delta_hidden
    b1 = b1 + lr * delta_hidden
```

```python
import matplotlib.pyplot as plt

plt.plot(loss)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss vs Epochs")
plt.show()
```

```python
test_x = np.array([[3]])
z1 = test_x * W1 + b1
a1 = sigmoid(z1)
z2 = a1 * W2 + b2
y_test_pred = sigmoid(z2)
print("Predicted Output:", y_test_pred)
```
